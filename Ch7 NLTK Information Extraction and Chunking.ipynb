```python
import nltk
nltk.data.path.append("/content/drive/My Drive/Colab Notebooks/nltk_data/")
```

```python
import  re, pprint

def ie_preprocess(document):
    sentences = nltk.sent_tokenize(document) 
    sentences = [nltk.word_tokenize(sent) for sent in sentences] 
    sentences = [nltk.pos_tag(sent) for sent in sentences]
    return sentences
    

doc1 = 'The fourth Wells account moving to another agency is the packaged paper-products division \
 of Georgia-Pacific Corp., which arrived at Wells only last fall. Like Hertz and the History Channel, \
 it is also leaving for an Omnicom-owned agency, the BBDO South unit of BBDO Worldwide. \
 BBDO South in Atlanta, which handles corporate advertising for Georgia-Pacific, will assume additional \
 duties for brands like Angel Soft toilet tissue and Sparkle paper towels, said Ken Haldin, a spokesman \
 for Georgia-Pacific in Atlanta.'
 
print(doc1)
```

```python
pprint.pprint(nltk.sent_tokenize(doc1))
```

```python
sentences = nltk.sent_tokenize(doc1) 
sentences = [nltk.word_tokenize(sent) for sent in sentences] 

pprint.pprint(sentences)
```

```python
pprint.pprint(ie_preprocess(doc1))
```

```python
ll = ie_preprocess(doc1)
ll
```

```python
#flattening
doc11 = [item for sublist in ll for item in sublist]
print(doc11)
```

### Using a regex-based grammar for extraction

```python
grammar = "myNP: {<DT>?<JJ>*<NN>}"
cp = nltk.RegexpParser(grammar)
result = cp.parse(doc11)

print(result)
```

```python
result.draw() # doesn't work on colab but works on a personal computer
```

### Another example

```python
grammar = "RELNP: {<NNP>+<IN><NNP>}"
cp = nltk.RegexpParser(grammar)
result = cp.parse(doc11)

print(result)
```

### A bigger grammar

```python

grammar = r"""
  NP: {<DT|JJ|NN.*>+}          # Chunk sequences of DT, JJ, NN
  PP: {<IN><NP>}               # Chunk prepositions followed by NP
  VP: {<VB.*><NP|PP|CLAUSE>+$} # Chunk verbs and their arguments
  CLAUSE: {<NP><VP>}           # Chunk NP, VP
  """


```

```python
cp = nltk.RegexpParser(grammar)

result = cp.parse(doc11)

print(result)
```

```python
result.draw()
```
