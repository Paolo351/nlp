```python
def gender_features(word):
    return {'last_letter': word[-1]}
print("gender_features('Shrek'): ",gender_features('Shrek'))
#{'last_letter': 'k'}

 
```

```python
from nltk.corpus import names
labeled_names = ([(name, 'male') for name in names.words('male.txt')] + \
[(name, 'female') for name in names.words('female.txt')])
```

```python
import random
random.shuffle(labeled_names)

print(labeled_names[0:50])
```

```python
featuresets = [(gender_features(n), gender) for (n, gender) in labeled_names]
train_set, test_set = featuresets[1000:], featuresets[:1000]

print("1: ",featuresets[0:50])
```

### Naive Bayes Classifier

```python
classifier = nltk.NaiveBayesClassifier.train(train_set)
print("classifier.classify(gender_features('Neo')",classifier.classify(gender_features('Neo')))
#'male'
```

```python
print( classifier.classify(gender_features('Trinity')) )
```

```python
print( classifier.classify(gender_features('Matrix')) )
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```

```python
def gender_features0(word):
    return {'first_letter': word[0]}
print("gender_features0('Shrek'): ",gender_features0('Shrek'))
```

```python
featuresets0 = [(gender_features0(n), gender) for (n, gender) in labeled_names]
train_set, test_set = featuresets0[1000:], featuresets0[:1000]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
#print("len(featuresets0)",len(featuresets0))
print(nltk.classify.accuracy(classifier, test_set))
```

```python
featuresets00 = featuresets + featuresets0
random.shuffle(featuresets00) 
print(featuresets00[0:50])

print("len(featuresets00)",len(featuresets00))
```

```python
train_set, test_set = featuresets00[1000:], featuresets00[:1000]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```

```python
def gender_features1(word):
    return {'almostlast_letter': word[-2]}
print("gender_features1('Shrek'): ",gender_features1('Shrek'))

featuresets1 = [(gender_features1(n), gender) for (n, gender) in labeled_names]
train_set, test_set = featuresets1[1000:], featuresets1[:1000]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```

```python
featuresets010 = featuresets00 + featuresets1
random.shuffle(featuresets010)
print(featuresets010[0:50])
 
print("len(featuresets010)",len(featuresets010))

train_set, test_set = featuresets010[1000:], featuresets010[:1000]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print('NB ', nltk.classify.accuracy(classifier, test_set))
```

```python
def gender_features2(word):
    return {'almostlast_letter': len(word)}
print("gender_features2('Shrek'): ",gender_features2('Shrek'))

featuresets2 = [(gender_features2(n), gender) for (n, gender) in labeled_names]
train_set, test_set = featuresets2[1000:], featuresets2[:1000]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```

```python
featuresets02 = featuresets00 + featuresets1 + featuresets2
random.shuffle(featuresets02)
print(featuresets010[0:50])
 
print("len(featuresets02)",len(featuresets02))

train_set, test_set = featuresets02[1000:], featuresets02[:1000]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print('NB ', nltk.classify.accuracy(classifier, test_set))
```
