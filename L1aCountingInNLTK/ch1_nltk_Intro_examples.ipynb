
#Counting words, and observing their distribution#

**Defining distributions for four texts**

```python
from nltk.book import FreqDist

fdist1 = FreqDist(text1)
fdist2 = FreqDist(text2)
fdist3 = FreqDist(text3)
fdist4 = FreqDist(text4)
```

**Counting occurences**

```python

print ("How many times does monstrous appear in Moby Dick: FreqDist text1 monstrous")
print (fdist1["monstrous"])
```

**Finding common contexts**

```python
print ("common contexts: finding the same neighbors")
print ('text2.common_contexts(["day", "night"])')
print (text2.common_contexts(["day", "night"]))
print ("\n")
print ('text2.common_contexts(["white", "pale"])')
print (text2.common_contexts(["white", "pale"]))

print ('text2.common_contexts(["a", "the"])')
text2.common_contexts(["a", "the"])

print ("\n")
print ('text4.common_contexts(["birth","death"])')
print (text4.common_contexts(["birth","death"]))
print ("end common contexts") 
```
