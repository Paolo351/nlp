```python

from nltk.book import FreqDist
from nltk.book import (text8)
#from nltk.corpus import webtext
#from nltk.util import bigrams
```

```python
currentText = text8
#currentText = webtext

print("currentText.collocations():   ")
currentText.collocations()
```

```python
fd_currentText = FreqDist(currentText)
ct_currentText = len(currentText)

fthe= fd_currentText['the']
fred= fd_currentText['red']
fwine= fd_currentText['wine']

fnever= fd_currentText['never']
fmarried= fd_currentText['married']
fis= fd_currentText['is']

print('word count', ct_currentText)
print('the', fthe)
print('red',fred)
print('wine',fwine)
print('married',fmarried)
print('never', fnever)
```
