```python
import re
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]

phoneWords=[w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)]
print(phoneWords)
```

```python
phoneWords=[w for w in wordlist if re.search('^[a-z]+[def][ghi][jlk][mno]$', w)]
print(phoneWords)

#print(wordlist)
goldwords = [w for w in wordlist if re.search('gold', w)]
print(goldwords)

```

```python
p = re.compile("[a-z]")
for m in p.finditer('a1b2c3d4'):
    print( m.start(), m.group())
    
```
