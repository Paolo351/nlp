## POS tagging in NLTK

```python
from nltk import wordpunct_tokenize as word_tokenize

text = word_tokenize("And now for something completely different")
print(nltk.pos_tag(text))
```

```python
text = word_tokenize("The more you prepare the better you do")
print(nltk.pos_tag(text))
"""
[('And', 'CC'), ('now', 'RB'), ('for', 'IN'), ('something', 'NN'),
('completely', 'RB'), ('different', 'JJ')]
"""
```

```python
text = word_tokenize("They didn't refuse to permit us to obtain the refuse permit")
print(nltk.pos_tag(text))

print(nltk.pos_tag(text,tagset='universal'))
```

 ### Finding words in a category

```python
wsj = nltk.corpus.treebank.tagged_words(tagset='universal')
word_tag_fd = nltk.FreqDist(wsj)
li = [wt[0] for (wt, _) in word_tag_fd.most_common() if wt[1] == 'VERB']
print("VERB",li[0:100])
```

```python
li = [wt[0] for (wt, _) in word_tag_fd.most_common() if wt[1] == 'NOUN']
print('NOUN',li[0:100])
```

```python
li = [wt[0] for (wt, _) in word_tag_fd.most_common() if wt[1] == 'ADJ']
print('ADJ',li[0:100])
```

```python
wsj = nltk.corpus.treebank.tagged_words(tagset='wsj')
word_tag_fd = nltk.FreqDist(wsj)
li = [wt[0] for (wt, _) in word_tag_fd.most_common() if wt[1] == 'VBZ']
print("VBZ",li[0:100])
```

```python
li = [wt[0] for (wt, _) in word_tag_fd.most_common() if wt[1] == 'NN']
print('NN',li[0:100])
```

```python
li = [wt[0] for (wt, _) in word_tag_fd.most_common() if wt[1] == 'JJ']
print('JJ',li[0:100])
 
```

## Distributional Similarity 

The text.similar() method takes a word w, finds all contexts w1 w w2, then finds all words w' that appear in the same context, i.e. w1 w' w2.

```python
text = nltk.Text(word.lower() for word in nltk.corpus.brown.words())
print("Distributional Similarity")
 
st = text.similar('woman')

st
```
